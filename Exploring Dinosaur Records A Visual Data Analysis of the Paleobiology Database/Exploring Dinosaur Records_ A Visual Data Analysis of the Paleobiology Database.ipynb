## **Import necessary Python libraries and Datasets.**

```python
# Import the packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dinosaurs = pd.read_csv('/content/dinosaurs.csv')
```

```python
# Preview the dataframe
dinosaurs
```

### **1.1. Analyzing the distribution of key features like dinosaur class and average size. Using histograms to visualize the distributions.**

```python
# create the histogram for dinosaur class

plt.figure(figsize=(6, 5))
plt.hist(data=dinosaurs, x='class')
plt.xlabel('Class')
plt.ylabel('Number of Dinosaurs')
plt.title('Distribution of Dinosaurs by Class')
plt.show()
```

```python
# Create a histogram for dinosaur average size
plt.figure(figsize=(8, 5))
plt.hist(data=dinosaurs, x='length_m')
plt.xlabel('Average Size (m)')
plt.ylabel('Number of Dinosaurs')
plt.title('Distribution of Dinosaurs by Average Size')
plt.show()
```

### **1.2. Creating a visualization (bar chart) to display the number of dinosaurs per type.**

```python
# Count the number of occurrences for each dinosaur type
type_counts = dinosaurs['type'].value_counts()

# Create a bar chart of dinosaur type counts with enhanced visualization
plt.figure(figsize=(12, 8))
plt.bar(type_counts.index, type_counts.values, color='skyblue')
plt.xlabel('Dinosaur Type', fontsize=14)
plt.ylabel('Number of Occurrences', fontsize=14)
plt.title('Number of Dinosaurs by Type', fontsize=16)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

The provided code block counts occurrences of each dinosaur type in a dataset and visualizes this data with a bar chart using `matplotlib.pyplot`. It highlights the most common dinosaur type and displays the distribution of types with enhanced chart features.

### **2.1. Analying the trend of dinosaur discoveries over time by using line plot to visualize the trend.**

```python
# Extract the relevant columns for analysis
discovery_dates = dinosaurs['max_ma']

# Calculate the number of discoveries per million years
discovery_counts = discovery_dates.value_counts().sort_index(ascending=True)

# Create a line plot of discovery counts over time
plt.figure(figsize=(10, 6))
plt.plot(discovery_counts.index, discovery_counts.values)
plt.xlabel('Time (Million Years Ago)')
plt.ylabel('Number of Discoveries')
plt.title('Trend of Dinosaur Discoveries Over Time')
plt.grid(True)
plt.show()
```

### **2.2. Identifying the most active types for dinosaur discoveries, by using histogram to visualize the distribution of discovery dates.**

```python
# Extract the relevant columns for analysis
discovery_types = dinosaurs['type']

# Calculate the number of discoveries per type
type_counts = discovery_types.value_counts().sort_index(ascending=True)

# Create a bar plot of discovery counts per type
plt.figure(figsize=(10, 6))
plt.bar(type_counts.index, type_counts.values)
plt.xlabel('Dinosaur Type')
plt.ylabel('Number of Discoveries')
plt.title('Most Active Types for Dinosaur Discoveries')
plt.grid(True)
plt.show()
```

### **2.4. Analyzing fossil records over time. By using bar chart plotting the number of fossil per million years to identify any trends.**
